```python
import pandas as pd
import numpy as np
```

```python
#Load file interval_data.xlsx using pandas package
df = pd.read_excel("C:\\Users\\LocalSystem\\Downloads\\Python_test\\Python_test\\interval_data.xlsx")
df.head()
```

```python
#transpose table so that it has columns such as MPAN, Date, Hour (values found in Header row from 00:00 to 23:30), Value (for example 7.18, 7.47 and so on). 
df_melted = pd.melt(df, id_vars=['MPAN', 'Date'], var_name='Hour', value_name='Value')
df_melted
```

```python
#Calculate mean, max, min value for each MPAN for each week

df_melted['Date'] = pd.to_datetime(df_melted['Date'])
result = df_melted.groupby(['MPAN', df_melted['Date'].dt.isocalendar().week])['Value'].agg(['mean', 'max', 'min'])
result
```

```python
# Save result to separate file for each MPAN number
for mpan, data in result.groupby('MPAN'):
    data.to_excel(f'{mpan}.xlsx')
```
